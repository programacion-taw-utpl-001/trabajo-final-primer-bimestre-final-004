## Libreria
### Pandas
Pandas es un paquete de Python que proporciona estructuras de datos similares a los dataframes de R. Pandas depende de Numpy, la libreria que añade un potente tipo matricial a Python. Los principales tipos de datos que puede representarse con pandas son:
* Datos tabulares con culmnas de tipo heterogéneo con etiquetas en columnas y filas.
* Series temporales.

Pandas proporciona herramientas que permiten:
* Leer y escribir datos en diferentes formatos: CSV, Microsoft Excel, bases SQL y formatos HDF5.
* Seleccionar y filtrar de manera sencilla tablas de datos en funcion de posición, valor o etiquetas.
* Fusionar y unir datos.
* Transformar datos aplicando funciones tanto en globas como por venta.
* Manipulación de series temporales
* Hacer gráficas

### Bokeh
Bokeh es un libreria interactiva de visualización de datos que se orienta a los navegadores web para presentaciones de datos. Bokeh como logro primordial es que otorga la capacidad de construir gráficas elegantes, consistentes e interactivas.

### Generar graficos:

Para visualizar la información podemos recurir a la utilización de graficas como herramienta de analicis de información, trabajando a partir de la filtración de los datos.

Este ejemplo de grádica es una generada a partir de una plantilla de la pagina de [Bokeh](https://bokeh.pydata.org/en/latest/docs/gallery.html) que proporsiona vario modelos.

```python
# importamos las librerias de bokeh para  que nos permita  graficar 
from bokeh.io import show
from bokeh.plotting import figure
```

```python
from bokeh.io import output_notebook, output_file
output_notebook()
```

```python
# graficamos  los valores de edades de mujeres de la provincia de loja
# que sean solteras y con un nivel de estudio de primaria del sector urbano
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6
from bokeh.plotting import figure
from bokeh.transform import factor_cmap

output_file("mujeres de loja.html")

fruits = ['20-25', '25-30', '30-35', '35-40', '40-80']
counts = [14, 14, 6, 5, 22, ]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, plot_height=350, toolbar_location=None, title="intervalo de mujeres solteras de loja")
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 9
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```

```python
# filtramos  una nueva tabla con los parametros de provinvia 'Loja', estado civil 'Viuda' y area 'Urbana'
tabla2 = tabla1[tabla1['prov_insc'] == 'Loja']
tabla3=tabla2[tabla2['est_civim']== 'Soltera']
tabla4=tabla3[tabla3['area_muj']=='Urbana']
tabla4
```

```python
# graficamos  los valores de los niveles de estudios  de muejeres solteras del sector urbano
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6
from bokeh.plotting import figure
from bokeh.transform import factor_cmap

output_file("bar_colormapped.html")

fruits = ['Sperior', 'Primaria', 'secundaria', 'PstBachillerato', 'Educacion Media']
counts = [14, 12, 20, 2, 7]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, plot_height=350, toolbar_location=None, title="Intervalo de nivel de educacion Mujeres")
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 9
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```

```python
# filtramos  una nueva tabla con los parametros de provinvia 'Loja' y estado civil 'Soltera'
tabla2 = tabla1[tabla1['prov_insc'] == 'Loja']
tabla3=tabla2[tabla2['est_civim']== 'Soltera']
tabla3
```

```python
# graficamos  los valores de los sectores Rurales y Urbanos de mujeres solteras
from bokeh.io import show, output_file
from bokeh.plotting import figure

output_file("bar_basic.html")

fruits = ['Rural', 'Urbano']

p = figure(x_range=fruits, plot_height=350, title="Intervalo de Sectores de Mujeres Solteras",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=[9, 50], width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
# graficamos  los valores de los sectores Rurales y Urbanos de mujeres viudas
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6
from bokeh.plotting import figure
from bokeh.transform import factor_cmap

output_file("bar_colormapped.html")

fruits = ['Urbano', 'Rural']
counts = [25, 2]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, plot_height=350, toolbar_location=None, title="Intervalo de nivel de viudas en sectores")
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 9
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```

```python
#procedemos a graficar itervalos de meses.
from bokeh.io import show, output_file
from bokeh.plotting import figure

output_file("bar_basic.html")

fruits = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo ','Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

p = figure(x_range=fruits, plot_height=350, title="Intervalo de mes de Incripcion de mujeres solteras.",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=[0, 3, 5, 4, 2, 7, 5, 10, 6, 8, 2, 6], width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```
